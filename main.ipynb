```python
import pandas as pd 
import numpy as np
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
import nltk
nltk.download('stopwords')
stopwords.words("english")
```

```python
df = pd.read_csv(r"C:\Users\pj183\Documents\me.python\python projects\machine learning\fakeNewsPrediction\train.csv")
df.shape
df
```

```python
df.isnull().sum()
```

replace the null values by empty strings

```python
new_df=df.fillna(" ")
```

```python
new_df['content']=new_df['author']+" "+new_df['title']
new_df.drop(columns=['title','author'],axis=1,inplace=True)
x = new_df.drop(columns='label')
y = new_df['label']
```

```python
# stemming - reducing the word to its root word
# actor,actress,acting -> act
```

```python
port_Stem=PorterStemmer()
```

```python
def stemm(content):
    stemmed = re.sub('[^a-zA-Z]',' ',content)
    stemmed = stemmed.lower()
    stemmed = stemmed.split()
    stemmed = [port_Stem.stem(word) for word in stemmed if not word in stopwords.words("english")]
    stemmed=' '.join(stemmed)
    return stemmed   
       
```

```python
new_df['content']=new_df['content'].apply(stemm )
```

```python
new_df
```

```python
x = new_df['content'].values
y= new_df['label'].values
x.shape,y.shape
```

```python
vectorizer = TfidfVectorizer()
vectorizer.fit(x)
x=vectorizer.transform(x)
```

train test split

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,stratify=y,random_state=42)
```

```python
model=LogisticRegression()
model.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import accuracy_score
ypred = model.predict(xtest)
accuracy=accuracy_score(ypred,ytest)
accuracy
```

```python
def predict_news(news_text):
    # Preprocess the input text
    processed_text = stemm(news_text)
    
    # Vectorize the input text
    vectorized_input = vectorizer.transform([processed_text])
    
    # Make prediction
    prediction = model.predict(vectorized_input)
    
    return "Fake News" if prediction[0] == 0 else "Real News"
```

```python
user_input = " Apple iPhone 17 Air launch in September seems more likely than ever"
result = predict_news(user_input)
print(f"The entered news is: {result}")
```
